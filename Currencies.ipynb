### Расчет курса рубля к доллару

```python
x = AssetList(['RUB.FOREX'])
```

```python
y = -x.ror
y = y.squeeze()
y.rename('RUB-USD', inplace=True)
wealth = 1000 * (1 + y).cumprod() - 1
wealth.plot(legend='RUB-USD', figsize=(12,6))
```
